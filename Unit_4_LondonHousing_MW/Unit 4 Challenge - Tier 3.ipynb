```python
# Now what is happening is the unnames are separators.  And between unknown 34 to city of london there is 32 identifers
# this is probably our 32 identifier that we are looking for....  Lets get their index and use iloc to get our london
# only datafrane
```

```python
import re 
index_london_city = []
for idx, id in enumerate(list(identifier.values.astype(str))):
    if re.search('E09*', id):
        index_london_city.append(idx - 1 ) # i am adding city of london for fun.  Also i turn Date into index so need -1 here
print(index_london_city)
```
