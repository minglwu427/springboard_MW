```python
main_df = pd.read_csv('millenniumofdata_v3_headlines.csv', index_col=0)
unit = main_df.iloc[0,:]
main_df = main_df.drop('Units', axis = 0)
main_df
```

**Ok, now lets only look at Modern Era.  The time line goes back way too far doesn't really have any meaning for us**

```python
main_df = main_df.loc['2000':'2016']
main_df = main_df.apply(lambda x :pd.to_numeric(x))
main_df.corr().round(2).style.background_gradient(cmap='coolwarm')
```

**Base on the result. We can see there is a lot of correlation between THE GDP with many blue including population and employment**
